```python
import pandas as pd
import numpy as np
```

# GDP

```python
gdp = pd.read_csv('GDP.csv')

gdp_melt = gdp.melt(id_vars = ['Country Name', 'Country Code'],
                    var_name='Year',
                    value_name = 'GDP growth % (1960 - 2018)')

gdp_pivot = pd.pivot_table(gdp_melt, values='GDP growth % (1960 - 2018)',
                           index = ['Country Name', 'Country Code'])

gdp_pivot = gdp_pivot.reset_index()


gdp_pivot
```

# Above GDP Mean Value

```python
gdp_pivot['GDP growth % (1960 - 2018)'].mean()
```

Countries above GDP mean value

```python
gdp_great = gdp_pivot[gdp_pivot['GDP growth % (1960 - 2018)']>=(gdp_pivot['GDP growth % (1960 - 2018)'].mean())]
```

```python
gdp_great
```

# GDP per capita 

```python
gdp_per_capita = pd.read_csv('GDP_per_capita.csv')

gdp_per_capita.head()

gdp_per_capita_melt = gdp_per_capita.melt(id_vars = ['Country Name', 'Country Code'],
                                          var_name='Year',
                                          value_name = 'GDP_PER_CAPITA growth % (1960 - 2018)')

gdp_per_capita_pivot = pd.pivot_table(gdp_per_capita_melt, values='GDP_PER_CAPITA growth % (1960 - 2018)',
                                      index = ['Country Name', 'Country Code'])

gdp_per_capita_pivot = gdp_per_capita_pivot.reset_index()


gdp_per_capita_pivot

gdp_per_capita_pivot['GDP_PER_CAPITA growth % (1960 - 2018)'].mean()

gdp_per_capita_great = gdp_per_capita_pivot[gdp_per_capita_pivot['GDP_PER_CAPITA growth % (1960 - 2018)']>=(gdp_per_capita_pivot['GDP_PER_CAPITA growth % (1960 - 2018)'].mean())]

gdp_per_capita_great
```

# Inflation, GDP deflator (annual %)

Inflation, GDP deflator (annual %)
Inflation as measured by the annual growth rate of the GDP implicit deflator shows the rate of price change in the economy as a whole. The GDP implicit deflator is the ratio of GDP in current local currency to GDP in constant local currency.

Definition: Inflation as measured by the annual growth rate of the GDP implicit deflator shows the rate of price change in the economy as a whole. The GDP implicit deflator is the ratio of GDP in current local currency to GDP in constant local currency.

```python
Inflation = pd.read_csv('INFLATION.csv')

Inflation.head()

Inflation_melt = Inflation.melt(id_vars = ['Country Name', 'Country Code'],
             var_name='Year',
             value_name = 'INFLATION growth % (1960 - 2018)')

Inflation_pivot = pd.pivot_table(Inflation_melt, values='INFLATION growth % (1960 - 2018)',
                           index = ['Country Name', 'Country Code'])

Inflation_pivot = Inflation_pivot.reset_index()

Inflation_pivot

Inflation_pivot['INFLATION growth % (1960 - 2018)'].mean()

Inflation_great = Inflation_pivot[Inflation_pivot['INFLATION growth % (1960 - 2018)']<=(Inflation_pivot['INFLATION growth % (1960 - 2018)'].mean())]

Inflation_great



```

# Population growth (annual %)

```python
population = pd.read_csv('population_growth_rate.csv')

Population_melt = population.melt(id_vars = ['Country Name', 'Country Code'],
                                  var_name='Year',
                                  value_name = 'Population growth annual-% (1960 - 2018)')

Population_pivot = pd.pivot_table(Population_melt, values='Population growth annual-% (1960 - 2018)',
                                  index = ['Country Name', 'Country Code'])

Population_pivot = Population_pivot.reset_index()

Population_great = Population_pivot[Population_pivot['Population growth annual-% (1960 - 2018)']>=(Population_pivot['Population growth annual-% (1960 - 2018)'].mean())]

Population_great


```

# Life expectancy at birth

```python
life_expectancy = pd.read_csv('Life expectancy at birth.csv')


Life_expectancy_melt = life_expectancy.melt(id_vars = ['Country Name', 'Country Code'],
                                            var_name='Year',
                                            value_name = 'Life expectancy at birth in Years(1960 - 2018)')

Life_expectancy_pivot = pd.pivot_table(Life_expectancy_melt, values='Life expectancy at birth in Years(1960 - 2018)',
                                       index = ['Country Name', 'Country Code'])

Life_expectancy_pivot = Life_expectancy_pivot.reset_index()

Life_expectancy_great = Life_expectancy_pivot[Life_expectancy_pivot['Life expectancy at birth in Years(1960 - 2018)']>=(Life_expectancy_pivot['Life expectancy at birth in Years(1960 - 2018)'].mean())]

Life_expectancy_great
```
